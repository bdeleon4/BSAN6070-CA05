```
#import packages 

import pandas as pd
import numpy as np
from sklearn import linear_model
import sklearn.metrics as metrics
import matplotlib.pyplot as plt
from sklearn.neighbors import NearestNeighbors
```

```
#read in data
data = pd.read_csv('https://github.com/ArinB/CA05-kNN/raw/master/movies_recommendation_data.csv')

#look at first few records
data.head(10)
```

```
new_data = data.copy()[['Biography', 'Drama','Thriller','Comedy','Crime', 'Mystery', 'History']]
new_data['new_rating'] = data.copy()['IMDB Rating']/10

#divided the rating by 10 so the range would minimize
```

```
new_data.head()
```

```
knn = NearestNeighbors(n_neighbors=5, algorithm='kd_tree').fit(new_data)
distances, indices = knn.kneighbors(new_data)
```

```
#we need to recommend 5 similar movies
def euclidean_distance(p1, p2):
    sum = 0
    len_p1 = len(p1) 
    for i in range(len_p1):
        sq_distance = (p1[i] - p2[i]) ** 2
        sum = (sum + sq_distance)
    return sum
```

```
#Setting up The post movie for recommendations

#IMDB Rating = 7.2
#Biography= Yes 
#Drama = Yes 
#Thriller = No 
#Comedy = No 
#Crime = No 
#Mystery = No 
#History = Yes

thepost = (1,1,0,0,0,0,1,0.72)
distances_n, indices_n = knn.kneighbors([thepost])
```

##Movie Recommender

```
distance=[]
movie_index=[]

for i in range(len(data)):
  new = euclidean_distance(thepost, new_data.iloc[i,:])

  distance.append(new)
  movie_index.append(i)
```

```
movies = pd.Series(distance,index=movie_index).sort_values(ascending=True)[:5]
recommendations = data.iloc[movies.index]

movie_rec = recommendations[['Movie Name']]
```
